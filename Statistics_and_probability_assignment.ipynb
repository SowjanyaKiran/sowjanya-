**2. Define the Central Limit Theorem and discuss its significance in statistical inference.**

**Ans:** The central limit theorem states that, given a sufficiently large sample size, the sampling distribution of the mean for a variable will approximate a normal distribution regardless of that variable’s distribution in the population.

image.png

**Significance in Statistical Inference**

The significance of the Central Limit Theorem in statistical inference lies in its ability to simplify analysis and facilitate hypothesis testing.

1. **Normality Assumption:** Many statistical methods and tests assume that data follows a normal distribution. The CLT allows researchers to apply these methods even when the underlying population does not meet this assumption, provided that the sample size is large enough 24.

2. **Confidence Intervals:** The CLT enables the construction of confidence intervals for estimating population parameters. By knowing that sample means will be normally distributed, statisticians can calculate ranges within which they expect the true population mean to fall 13.

3. **Hypothesis Testing:** The CLT underpins many hypothesis tests, such as t-tests and z-tests. It allows for determining whether observed differences between groups are statistically significant by comparing sample means to hypothesized population means 56.

4. **Applicability Across Disciplines:** The theorem is widely applicable across various fields, including economics, psychology, and healthcare, making it a cornerstone of empirical research.



**3. Discuss the concept of sampling and its role in statistical analysis**

**Ans:** Sampling is a key concept in statistics where a small group (sample) is selected from a larger population to draw conclusions about the whole population. Since it’s often too expensive or time-consuming to gather data from everyone, sampling makes the process easier and more manageable.

**Types of Sampling**

**Probability Sampling:**
In probability sampling, everyone in the population has a known chance of being selected. This helps ensure the sample is fair and representative of the population. Common methods include:

**1. Simple Random Sampling:** Every individual has an equal chance of being chosen, like drawing names from a hat.

**2. Stratified Sampling:** The population is divided into groups (e.g., age or gender), and samples are randomly chosen from each group to ensure all groups are represented.

**3. Cluster Sampling:** The population is divided into groups (often by location), and whole groups are randomly selected.

**Non-Probability Sampling:**
In non-probability sampling, some individuals may not have a chance of being chosen, which can introduce bias. These methods are easier and cheaper but less reliable.

**1. Convenience Sampling:** Samples are chosen based on who’s easiest to access, like surveying people at a mall.
Judgmental Sampling: The researcher selects individuals based on their own judgment.

**2. Quota Sampling:** Researchers ensure certain groups are represented by selecting a specific number of individuals from each group.
Role of Sampling in Statistical Analysis

**Sampling is important because:**

**1. Feasibility:** Its not always possible to gather data from everyone, so sampling makes it easier and cheaper.

**2. Generalization:** By studying a sample, researchers can make conclusions about the whole population.

**3. Bias Reduction:** Probability sampling methods help ensure fairness and reduce bias.

**4. Statistical Testing:** Proper sampling is needed to run tests and make predictions based on the data.

**4. Explain the process of hypothesis testing and the key components involved.**

**Ans:** Hypothesis testing is a statistical method that helps us decide if there is enough evidence in a sample of data to support or reject a claim about a larger group (population). It allows researchers to make informed conclusions based on the sample data they collect.

Hypothesis testing is a fundamental method in statistics used to make inferences about a population based on sample data. It involves evaluating two competing hypotheses: the null hypothesis (H₀) and the alternative hypothesis (H₁).

4237247_1626434645_HYPOTHESIS TESTINGArtboard 1.jpg

**Key Components of Hypothesis Testing**

**Null Hypothesis (H₀):**

A statement that assumes no effect, no difference, or no relationship in the population.
Example: The mean score of Group A is equal to the mean score of Group B.

**Alternative Hypothesis (H₁):**

A statement that contradicts the null hypothesis, suggesting there is an effect, difference, or relationship.
Example: The mean score of Group A is not equal to the mean score of Group B.

**Test Statistic:**

A numerical value calculated from the sample data, used to determine whether to reject the null hypothesis.
Common test statistics include the z-score, t-score, F-statistic, or chi-square statistic.

**Significance Level (α):**
The probability threshold for rejecting the null hypothesis, typically set at 0.05 (5%) or 0.01 (1%).
It represents the risk of rejecting the null hypothesis when it is true (Type I error).

**P-value:**

The probability of obtaining the observed test statistic (or more extreme) under the assumption that the null hypothesis is true.
If the p-value is less than α, the null hypothesis is rejected.

**Decision Rule:**

A rule based on the comparison of the p-value and α to decide whether to reject or fail to reject the null hypothesis.
Reject H₀ if p-value ≤ α; otherwise, fail to reject H₀.

**Conclusion:**

A statement summarizing the results of the hypothesis test in the context of the problem.

**Types of Hypothesis Tests**

**1. One-sample tests:** Compare a sample mean or proportion to a known value.

**2. Two-sample tests:** Compare means or proportions between two groups.

**3. Paired tests:** Compare measurements from the same subjects under different conditions.

**4. Chi-square tests:** Test for independence or goodness-of-fit.

**5. ANOVA:** Compare means across three or more groups.

 **5.Describe the T-distribution and how it differs from the normal distribution.**

The t-distribution is a probability distribution used in hypothesis testing for small sample sizes (typically less than 30) or when the population standard deviation is unknown.

t-distribution-vs-z-distribution.jpg

As the degrees of freedom (total number of observations minus 1) increases, the t-distribution will get closer and closer to matching the standard normal distribution, a.k.a. the z-distribution, until they are almost identical.

Above 30 degrees of freedom, the t-distribution roughly matches the z-distribution. Therefore, the z-distribution can be used in place of the t-distribution with large sample sizes.

The z-distribution is preferable over the t-distribution when it comes to making statistical estimates because it has a known variance. It can make more precise estimates than the t-distribution, whose variance is approximated using the degrees of freedom of the data.

**Applications of the T-Distribution**

**One-Sample T-Test:** Determines if the mean of a single sample differs from a known or hypothesized population mean.

**Independent Two-Sample T-Test:** Compares the means of two independent groups to see if they differ significantly.

**Paired Sample T-Test:** Compares means from the same group at different times (e.g., before and after treatment).




**6. Calculate the mean, median, and standard deviation for the following dataset: [10, 15, 20, 25, 30].**

The mean is the average of a set of numbers. It shows the central value of the data.

**How to Calculate the Mean:**

1.Add up all the numbers in the dataset.

2.Divide the total by the number of values.

**Median**
The median is the middle value in a set of numbers when they are arranged in ascending (small to large) or descending order. It shows the central point of the data.

1.Arrange the numbers in order.

2.Check the total number of values:

The mode is the number that appears most frequently in a dataset. It shows which value occurs the most.

1.List the numbers.

2.Count how many times each number appears.

```python
import numpy as np
from scipy import stats

# Data
data = [10, 15, 20, 25, 30]

# Mean
mean = np.mean(data)
print(f"Mean is {mean}")

# Median
median = np.median(data)
print(f"Median is {median}")

# Mode
mode_result = stats.mode(data)
# Access mode and count directly, without indexing
print(f"Mode is {mode_result.mode} (Count: {mode_result.count})")
```

**7.A researcher wants to estimate the average height of students in a university. She samples 50 students and finds the mean height to be 65 inches with a standard deviation of 3 inches. Construct a 95% confidence interval for the population mean height**

**Ans:** Given information mean height x̄=65 , 𝞼=3 Confidence interval (CI) =x̄ ± Z * (𝞼 ÷ √n)

Z=0.95 value=0.8264

 confidence_interval=[65-0.8289*(3/√50), 65+ 0.8289*(3/√50)

 confidence_interval=[65-0.293, 65+0.293]

 confidence_interval=[64.707,65.293]

```python
import numpy as np
from scipy import stats

mean = 65        # Sample mean
std = 3          # Standard deviation
n = 50           # Sample size
ci = 0.95        # Confidence level

# Step 1: Calculate the Z-critical value
z_critical = stats.norm.ppf(1 - (1 - ci) / 2)

# Step 2: Calculate the Margin of Error
margin_of_error = z_critical * (std / np.sqrt(n))

# Step 3: Calculate the Confidence Interval
confidence_interval = [mean - margin_of_error, mean + margin_of_error]

# Step 4: Display the Confidence Interval
print(confidence_interval)
```

**8.A manufacturer claims that the average lifespan of its light bulbs is 1000 hours. A random sample of 50 light bulbs has a mean lifespan of 980 hours with a standard deviation of 50 hours. Test the manufacturer's claim at a significance level of 0.05 using a right-tailed hypothesis test.**

**Ans:**
we then perform a hypothesis test using the following null and alternative hypotheses:

Null Hypothesis (H0): The average lifespan of the light bulbs is 1000 hours (μ=1000).

Alternative Hypothesis (H1): The average lifespan of the light bulbs is less than 1000 hours (μ > 1000).


```python
import numpy as np
import scipy.stats as stats

# Given data
sample_mean = 980          # (x̄)
population_mean = 1000     # (μ)
sample_std_dev = 50        # (s)
sample_size = 50           # (n)
alpha = 0.05               # Significance level

# Calculate the z-test statistic
z = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Critical z-value for a right-tailed test
critical_value = stats.norm.ppf(1 - alpha)

# Display results
print(f"Calculated z-value: {z:.4f}")
print(f"Critical z-value at alpha {alpha}: {critical_value:.4f}")

# Decision rule for right-tailed test
if z > critical_value:
    print("Reject the null hypothesis. There is sufficient evidence to conclude that the average lifespan is greater than 1000 hours.")
else:
    print("Do not reject the null hypothesis. There is not enough evidence to conclude that the average lifespan is greater than 1000 hours.")
```

**10.A quality control manager at a factory wants to ensure that the average weight of products coming off the production line is 500 grams. She takes a random sample of 30 products and finds the mean weight to be 495 grams with a standard deviation of 10 grams. Test the manager's claim at a significance level of 0.01 using a left-tailed hypothesis test**

**Ans:**
**Hypothesis Statements**
**Null Hypothesis (H0):** The average weight of the products is 500 grams (μ = 500).
**Alternative Hypothesis (Ha):** The average weight of the products is less than 500 grams (μ < 500).

```python
import numpy as np
import scipy.stats as stats

# Given data
sample_mean = 495          # (x̄) Sample mean
population_mean = 500      # (μ) Population mean under H₀
sample_std_dev = 10        # (s) Standard deviation
sample_size = 30           # (n) Sample size
alpha = 0.01               # Significance level for the test

# Step 1: Calculate the z-test statistic
z = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Step 2: Determine the critical z-value for a left-tailed test
critical_value = stats.norm.ppf(alpha)  # Since it's left-tailed, use alpha directly

# Step 3: Display results
print(f"Calculated z-value: {z:.4f}")
print(f"Critical z-value at alpha {alpha}: {critical_value:.4f}")

# Step 4: Apply the decision rule for a left-tailed test
if z < critical_value:
    print("Reject the null hypothesis. There is sufficient evidence to conclude that the average weight is less than 500 grams.")
else:
    print("Do not reject the null hypothesis. There is not enough evidence to conclude that the average weight is less than 500 grams.")
```
