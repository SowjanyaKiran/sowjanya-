```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = '/content/Iris.csv'
data = pd.read_csv(url)
data.head()
```

```python
#1. What is the distribution of sepal lengths and widths among the Iris species?
# Created a histogram to show the distribution of SepalLengthCm
sns.histplot(x='SepalLengthCm', data=data,bins=20, palette='viridis')
plt.title('Distribution of Sepal Lengths')
plt.xlabel('SepalLengthCm')
plt.ylabel('Count')
plt.show()
#I used a histogram because it’s ideal for showing the distribution of numerical data.The sepal lengths and widths are divided into 20 bins to show how the data is spread. The x-axis shows the SepalLengthCm, and the y-axis shows counts the Iris species.

# Created a separate histogram for SepalWidthCm
sns.histplot(x='SepalWidthCm', data=data,bins=20, palette='viridis')
plt.title('Distribution of Sepal Widths')
plt.xlabel('SepalWidthCm')
plt.ylabel('Count')
plt.show()
#I used a histogram because it’s ideal for showing the distribution of numerical data.The sepal lengths and widths are divided into 20 bins to show how the data is spread. The x-axis shows the SepalWidthCm, and the y-axis shows counts the Iris species.
```

```python
#2. How does the petal length vary across different Iris species?
data_petalength = data.groupby(["Species", "PetalLengthCm"])["PetalLengthCm"].count()
data_petalength.plot(kind="bar", color="y", legend=True, position=0, width=0.25, label="Petal Length")
plt.rcParams["figure.figsize"] = (30, 20)
plt.title('petal length vary across different Iris species')
plt.xlabel("Species")
plt.ylabel("#Petal Length")
#I used a bar plot here because it's a great way to show categorical data. In this case, we’re comparing the count data of different species and their petal lengths. The **x-axis** represents the species, and the **y-axis** shows how many times each petal length occurs for each species. This bar plot helps us easily compare the petal lengths across the different **Iris species**.
```

```python
#3. Is there a relationship between sepal length and petal length?
sns.relplot(x ="SepalLengthCm", y ="PetalLengthCm",  hue ="Species", data = data, palette='magma');
#sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm',hue ="Species", data=data, palette='magma')
plt.title('Relationship between sepal length and petal length')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.show()
#I used a scatter plot to show how sepal length and petal length are related. The x-axis shows the sepal length, and the y-axis shows the petal length. The colors represent different species, making it easy to see how each species' sepal and petal lengths are connected. The 'magma' palette adds color to make it clearer. This plot helps us understand if there's a pattern between the two measurements and how it differs for each species.
```

```python
#4. How does the petal width vary across different Iris species?
data_petalength = data.groupby(["Species", "PetalWidthCm"])["PetalWidthCm"].count()
data_petalength.plot(kind="bar", color="g", legend=True, position=0, width=0.55, label="petal width")
plt.rcParams["figure.figsize"] = (30, 20)
plt.title('petal width vary across different Iris species')
plt.xlabel("Species")
plt.ylabel("PetalWidthCm")
plt.show()
#I used a bar plot to show how petal width varies across different Iris species here because it's a great way to show categorical data.The x-axis represents the species, and the y-axis shows the count of each petal width. figure.figsize is used to set the size of the plot. It adjusts the dimensions of the figure to make it larger and easier to read.
```

```python
#5. Are there any patterns observable in the pairwise relationships between features(sepal length, sepal width, petal length, petal width)?
sns.pairplot(data=data, hue="Species", palette="Set2")
plt.show()
#I used a pairplot to show the pairwise relationships between the features (sepal length, sepal width, petal length, and petal width).A pairplot creates scatter plots for each combination of two features, helping us see how the features relate to one another.
```

```python
#6. Can we distinguish between Iris species based on sepal and petal measurements?
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=data,x="SepalLengthCm",y="SepalWidthCm",hue="Species",palette="Set1",s=100)
plt.title("Sepal Measurements")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.subplot(1, 2, 2)
sns.scatterplot(data=data,x="PetalLengthCm",y="PetalWidthCm",hue="Species",palette="Set2",s=100)
plt.title("Petal Measurements")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")

plt.tight_layout()
plt.show()
#I used a scatter plot to visualize the relationships between different sepal and petal measurements for the Iris species. In each plot, the x-axis represents one measurement (like sepal length or petal length), and the y-axis represents another (like sepal width or petal width).
```
