```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = '/content/Iris.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.tail()
```

**1.What is the distribution of sepal lengths and widths among the Iris species?**

```python
plt.figure(figsize=(12, 6))

# Plot Sepal Length distribution
sns.histplot(x='SepalLengthCm', data=data, bins=20, color='blue', label='Sepal Length', kde=True)

# Plot Sepal Width distribution
sns.histplot(x='SepalWidthCm', data=data, bins=20, color='green', label='Sepal Width', kde=True)

# Add labels and title
plt.title('Distribution of Sepal Lengths and Widths')
plt.xlabel('Measurement (cm)')
plt.ylabel('Count')
plt.legend()

# Show the plot
plt.show()
```

**2.How does the petal length vary across different Iris species?**

```python
# Plot bar chart for Petal Length distribution across Species
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="PetalLengthCm", hue="Species", multiple="stack", bins=20)
#multiple="stack" --> Stacks the bars for each category on top of each other.
#The total height of each bar represents the combined count of all categories at that bin.

# Add labels and title
plt.title("Petal Length Variation Across Iris Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Count")

# Show the plot
plt.show()
```

**3.Is there a relationship between sepal length and petal length?**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue="Species", data=data, palette='magma')

# Add labels and title
plt.title('Relationship Between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')

# Show plot
plt.show()
```

**4.Are there any patterns observable in the pairwise relationships between features (sepal length, sepal width, petal length, petal width)?**

```python
sns.pairplot(data=data, hue="Species", palette="Set2")
plt.show()
```

**5.Can we distinguish between Iris species based on sepal and petal measurements?**

```python
# Create figure
plt.figure(figsize=(12, 6))

# First subplot: Sepal Measurements
plt.subplot(1, 2, 1)
sns.scatterplot(data=data, x="SepalLengthCm", y="SepalWidthCm", hue="Species", palette="Set1", s=100)
plt.title("Sepal Measurements")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

# Second subplot: Petal Measurements
plt.subplot(1, 2, 2)
sns.scatterplot(data=data, x="PetalLengthCm", y="PetalWidthCm", hue="Species", palette="Set2", s=100)
plt.title("Petal Measurements")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")

# Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
url = '/content/diabetes.csv'
data = pd.read_csv(url)
data.head()
```

**1.What is the distribution of age and BMI (Body Mass Index) among diabetic patients?**

```python
plt.figure(figsize=(10, 6))

# Kernel Density Estimate Plot for Age and BMI distribution .It is smooth representation of a histogram that helps visualize the probability density of data.
sns.kdeplot(data=data, x="Age", color="blue", fill=True, label="Age")
sns.kdeplot(data=data, x="BMI", color="green", fill=True, label="BMI")

# Add labels and title
plt.title("Distribution of Age and BMI Among Diabetic Patients")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

**2.How do blood pressure levels vary across different age groups?**

```python
plt.figure(figsize=(20, 6))
sns.boxplot(x="Age", y="BloodPressure", data=data, palette="coolwarm")
plt.title("Blood Pressure Levels Across Different Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Blood Pressure")
plt.show()
```

**3.Is there a correlation between BMI and Glucose levels?**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="BMI", y="Glucose", data=data, hue="Outcome", palette="magma")
plt.title("Correlation Between BMI and Glucose Levels")
plt.xlabel("BMI")
plt.ylabel("Glucose Levels")
plt.show()
```

**4.Are there any observable patterns in the relationship between diabetes progression and patient age?**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="Outcome", y="Age", data=data, palette="viridis")
plt.title("Diabetes Progression and Age")
plt.xlabel("Diabetes Outcome (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

**5.How do insulin levels vary across different age groups?**

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x="Age", y="Insulin", data=data, palette="coolwarm")
plt.title("Insulin Levels Across Different Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Insulin Levels")
plt.show()
```
