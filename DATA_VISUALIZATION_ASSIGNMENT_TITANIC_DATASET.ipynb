```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = '/content/titanic.csv'
data = pd.read_csv(url)
data.head()
```

```python
#1. What is the distribution of passenger ages on the Titanic?
#example for histograms(Show the distribution of a single variable)
sns.histplot(data['Age'], bins=40, kde=True, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
#I used a histogram because it’s ideal for showing the distribution of numerical data. The passenger ages are divided into 40 bins to show how the data is spread. The x-axis shows the ages of passengers, and the y-axis shows how many passengers fall into each age group.
```

example for histograms(Show the distribution of a single variable)

```python
# 2. How does the survival rate vary based on passenger class (Pclass)?
sns.barplot(x='Pclass', y='Survived', data=data , palette='magma')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
#I used a bar plot to analyze how the survival rate varies by passenger class (Pclass) because it effectively compares the categorical variable Pclass with the numerical survival rate. The x-axis represents the passenger classes (1, 2, 3), while the y-axis shows the survival rate, calculated as the proportion of passengers who survived in each class.
```

```python
#3. Is there a relationship between the fare paid and the likelihood of survival?
sns.barplot(x='Survived', y='Fare', data=data , palette='Set2')
plt.title('Fare vs Survival Rate')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
#I used a bar plot to explore the relationship between the fare paid and the likelihood of survival because it effectively compares the categorical variable.The x-axis shows whether passengers survived (0 means they didn’t survive, 1 means they did), and the y-axis shows the average fare paid by each group.
```

```python
#4. How does the survival rate differ between male and female passengers?
sns.barplot(x='Sex', y='Survived', data=data , palette='Set2')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
#I used a bar plot to show how the survival rate differs between male and female passengers because it effectively compares the categorical data. The x-axis represents gender (male and female) and the y-axis shows the survival rate, which is the proportion of passengers who survived in each group
```

```python
#5 .What impact does the number of siblings/spouses (SibSp) or parents/children(Parch) aboard have on survival?
sns.boxplot(x='SibSp', y='Survived', data=data , palette='turbo')
plt.title('Survival Rate by Number of Siblings/Spouses (SibSp)')
plt.xlabel('SibSp')
plt.ylabel('Survival Rate')
plt.show()

sns.barplot(x='Parch', y='Survived', data=data, palette='turbo')
plt.title('Survival Rate by Number of Parents/Children (Parch)')
plt.xlabel('Parch')
plt.ylabel('Survival Rate')
plt.show()
#I used a box plot to show how the number of siblings/spouses (SibSp) affects survival. The x-axis represents the number of siblings/spouses, and the y-axis shows the survival rate for each group.
#I used a bar plot to show how the number of parents/children (Parch) affects survival. The x-axis shows the number of parents/children, and the y-axis shows the survival rate for each group.
```

```python
#6. Are there any significant differences in survival rates among passengers embarked from different ports?
sns.barplot(x='Embarked', y='Survived', data=data , palette='viridis')
plt.title('Survival Rate by Embarkation Port')
plt.xlabel('Embarked')
plt.ylabel('Survival Rate')
plt.show()
#I used a bar plot to compare the survival rates of passengers who embarked from different ports because it's a clear way to show differences between categories. The x-axis represents the different embarkation ports (C, Q, and S), and the y-axis shows the survival rate for passengers from each port.
```
